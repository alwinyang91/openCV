# 3. Render Image

```python
# !pip install matplotlib
```

```python
from matplotlib import pyplot as plt
```

```python
# Show image
plt.imshow(img)
```

```python
recolor = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(recolor)
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(gray.shape)
# Now gray is a 2D matrix in which there are no colors, only a "luminosity" value. So matplotlib applied a colormap by default.
plt.figure(figsize=(8,4))
plt.subplot(1, 2, 1)
plt.imshow(gray)
plt.subplot(1, 2, 2)
plt.imshow(gray, cmap="gray")
plt.show()
```

# 4. Write out Image

```python
output_path = os.path.join('data','images_output', 'outputgray.jpg')
```

```python
cv2.imwrite(output_path, gray)
```
