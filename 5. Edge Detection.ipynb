# 1. Import Dependencies

```python
import cv2
import os
```

# 2. Load Image

```python
image_path = os.path.join('data','images', 'bike.jpeg')
```

```python
img = cv2.imread(image_path)  # become a np array
img
```

```python
# Import matplotlib so we can visualize an image
from matplotlib import pyplot as plt
```

```python
plt.imshow(img)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# resize image
resized_img = cv2. resize(img, (int(img.shape[1]*2), int(img.shape[0]*2)))
cv2.imshow("Frame View", resized_img)
cv2.waitKey(1)
cv2.destroyAllWindows()
```

```python
cv2.waitKey(1)
cv2.destroyAllWindows()
```

# 3. Edge Detection

```python
#https://learnopencv.com/edge-detection-using-opencv/
```

```python
canny = cv2.Canny(img, threshold1=180, threshold2=200)  # you can customize the threshold
resized = cv2.resize(canny,  (int(img.shape[1]*2), int(img.shape[0]*2)))
cv2.imshow("Canny", resized)
cv2.waitKey(1)
cv2.destroyAllWindows()
```

```python
cv2.waitKey(1)
cv2.destroyAllWindows()
```

## remove noise with Gaussian filter 

```python
blurred = cv2.GaussianBlur(img, (5,5),0)  # you can customize the kernel size (3,3)
canny = cv2.Canny(blurred, threshold1=180, threshold2=200)
```

```python
resized = cv2.resize(canny,  (int(img.shape[1]*2), int(img.shape[0]*2)))
cv2.imshow("Canny", resized)
cv2.waitKey(1)
cv2.destroyAllWindows()
```

```python
cv2.waitKey(1)
cv2.destroyAllWindows()
```

## improve by change img to gray

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(img, (5,5),0)  # you can customize the kernel size (3,3)
canny = cv2.Canny(blurred, threshold1=180, threshold2=200)
resized = cv2.resize(canny,  (int(img.shape[1]*2), int(img.shape[0]*2)))
cv2.imshow("Canny", resized)
cv2.waitKey(1)
cv2.destroyAllWindows()
```

```python
cv2.waitKey(1)
cv2.destroyAllWindows()
```

# 4. Export Edges

```python
cv2.imwrite(os.path.join('data','images', 'bike-edges.jpg'), resized)
```
